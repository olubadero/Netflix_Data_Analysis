## Which year  had the most show addings by Netflix

```python
##create the table, insert an index and sort the years in ascending order
Yearly_Uploads = Netflix.Year.value_counts().reset_index().sort_values('index')
Yearly_Uploads
```

```python
## Rename the column names, order the index values, whilst dropping the old index numbering
Yearly_Uploads.columns = ['Year', 'Upload_per_Year']
Yearly_Uploads.reset_index(drop= True, inplace =True)
Yearly_Uploads
```

```python
plt.figure(figsize=(12, 6))
plt.xticks(rotation=0)
plt.title('Upload Count by Year')
plt.xlabel('Year')
plt.ylabel('Number of Uploads')
plt.plot(Yearly_Uploads.Year, Yearly_Uploads.Upload_per_Year, 'g-x', linewidth=4, markersize=12);
```

**From the visualization, it is clear that 2019 was the year that Netflix added the highest number of shows. Between 2016 and 2019, there were a lot of uploads by Netflix. However, this has continued to drastically reduced after 2019 until 2021. This may be due to COVID-19 pandemic, which may have affected film production and work productivity of Netflix staff**

## Which month had the most uploads by Netflix

```python
Netflix.Month.value_counts()
```

```python
## Replace the numbers with their monthly values so its easier understood
Netflix.Month.replace(1, 'January', inplace=True)
Netflix.Month.replace(2, 'February', inplace=True)
Netflix.Month.replace(3, 'March', inplace=True)
Netflix.Month.replace(4, 'April', inplace=True)
Netflix.Month.replace(5, 'May', inplace=True)
Netflix.Month.replace(6, 'June', inplace=True)
Netflix.Month.replace(7, 'July', inplace=True)
Netflix.Month.replace(8, 'August', inplace=True)
Netflix.Month.replace(9, 'September', inplace=True)
Netflix.Month.replace(10, 'October', inplace=True)
Netflix.Month.replace(11, 'November', inplace=True)
Netflix.Month.replace(12, 'December', inplace=True)
Netflix
```

```python
Monthly_Uploads = Netflix.Month.value_counts().reset_index()
Monthly_Uploads.columns = ['Month', 'Upload_per_Month']
Monthly_Uploads
```

```python
plt.figure(figsize=(15, 6))
plt.xticks(rotation=80)
plt.title('Monthly Netflix Upload')

sns.barplot(Monthly_Uploads.Month, Monthly_Uploads.Upload_per_Month);
```

**April has the highest monthly upload, whilst February has the lowest upload**

## Compare Upload type over the years

```python
Netflix.loc[(Netflix['Year']== 2021) & (Netflix['Month'] == 'September')]
```

```python
Movie = Netflix.loc[Netflix['type'] == 'Movie']
Movie
```

```python
TV_Show = Netflix.loc[Netflix['type'] == 'TV Show']
TV_Show
```

```python
Movies = Movie.Year.value_counts()
Movies = Movies.reset_index()
Movies.columns = ['Year', 'Upload_Count']
Movies.sort_values('Year', inplace= True)
Movies
```

```python
Movies.reset_index(drop=True, inplace=True)
Movies
```

```python
TV = TV_Show.Year.value_counts()
TV = TV.reset_index()
TV.columns = ['Year', 'Upload_Count']
TV.sort_values('Year', inplace= True)
TV
```

```python
TV.reset_index(drop=True, inplace=True)
TV
```

```python
plt.figure(figsize=(14, 6))
plt.xticks(rotation=0)
plt.title('Upload Type over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Uploads')
plt.plot(Movies.Year, Movies.Upload_Count, 'g-x', linewidth=4, markersize=12);
plt.plot(TV.Year, TV.Upload_Count, 'b-x', linewidth=4, markersize=12);
```

**From the above illustration it is clear that Movies has consistently dominated Netflixx uploads over the year. TV Shows were not even uploaded until 2013 and their upload has been very low and limited, Netflix may wish to consider investing in uploading more TV Shows.**
